```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
# import skleaarn pca module
from sklearn.decomposition import PCA

# The datasets are already preprocessed...
dataset1 = pickle.load(open("../datasets/part2_dataset_1.data", "rb"))
dataset2 = pickle.load(open("../datasets/part2_dataset_2.data", "rb"))
```

```python
from sklearn.cluster import KMeans

# create a kmeans model with 6 clusters
kmeans = KMeans(n_clusters=6)

# fit the kmeans model to the data
kmeans.fit(dataset1)

pca = PCA(n_components=2)

# fit the pca model to the data
pca.fit(dataset1)

# transform the data to the new basis
dataset1_pca = pca.transform(dataset1)

# plot the data in the new basis
plt.scatter(dataset1_pca[:,0], dataset1_pca[:,1], c=kmeans.labels_)
plt.title("PCA of dataset 1")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()


```

```python
%matplotlib inline
pca = PCA(n_components=3)

# fit the pca model to the data
pca.fit(dataset2)

kmeans = KMeans(n_clusters=3)

# fit the kmeans model to the data
kmeans.fit(dataset2)

# transform the data to the new basis
dataset2_pca = pca.transform(dataset2)

# plot the data in the new basis in 3d plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(dataset2_pca[:,0], dataset2_pca[:,1], dataset2_pca[:,2], c=kmeans.labels_)
plt.title("PCA of dataset 2")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
plt.show()








```
