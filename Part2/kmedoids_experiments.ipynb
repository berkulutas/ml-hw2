```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn_extra.cluster import KMedoids

from sklearn.metrics import silhouette_score
```

```python
# The datasets are already preprocessed...
# dataset1 = pickle.load(open("../datasets/part2_dataset_1.data", "rb"))
# dataset2 = pickle.load(open("../datasets/part2_dataset_2.data", "rb"))

dataset1 = pickle.load(open("../data2023/part2_dataset_1.data", "rb"))
dataset2 = pickle.load(open("../data2023/part2_dataset_2.data", "rb"))
```

```python
k_range = range(2, 11)

# Dataset 1
losses = []
silhouette_scores = []

for k in k_range:
    kmedoids = KMedoids(n_clusters=k, random_state=331).fit(dataset1)
    y_pred = kmedoids.predict(dataset1)
    silhouette_scores.append(silhouette_score(dataset1, y_pred))
    losses.append(kmedoids.inertia_)
    print(f"K={k}, Loss={kmedoids.inertia_}, Silhouette={silhouette_score(dataset1, y_pred)}")


plt.plot(k_range, losses)
plt.xlabel("K")
plt.ylabel("Loss")
plt.title("KMedoids Loss vs K (Dataset 1)")
plt.grid()
```

```python
plt.plot(k_range, silhouette_scores)
plt.xlabel("K")
plt.ylabel("Silhouette Score")
plt.title("KMedoids Silhouette Score vs K (Dataset 1)")
plt.grid()
```
